## 데이터 분석(EDA)

우선적으로 train data와 dev data를 통해 데이터의 분포를 알아보기

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_tr = pd.read_csv("../raw/train.csv")
df_v = pd.read_csv("../raw/dev.csv")
df_tt = pd.read_csv("../raw/test.csv")
```

```python
sns.histplot(df_tr["label"])
```

```python
sns.countplot(x="binary-label", data=df_tr)
```

```python
df_tr["len_sentence_1"] = df_tr["sentence_1"].apply(lambda x: len(x.split()))
df_tr["len_sentence_2"] = df_tr["sentence_2"].apply(lambda x: len(x.split()))

# 문장 길이 분포 시각화
sns.histplot(df_tr["len_sentence_1"], label="Sentence 1")
```

```python
sns.histplot(df_tr["len_sentence_2"], color="orange", label="Sentence 2")
```

```python
sns.scatterplot(x="len_sentence_1", y="len_sentence_2", hue="label", data=df_tr)
plt.title("Sentence Length vs Label")
```

```python
df_tr["source"].value_counts().plot(kind="bar")
```

```python
# 라벨값이 5인 데이터 확인
label_5_df = df_tr[df_tr["label"] == 5]

output_file = "./label_5_df.csv"
label_5_df.to_csv(output_file, index=False)

print(f"새로운 dataset {output_file} 생성")
```
