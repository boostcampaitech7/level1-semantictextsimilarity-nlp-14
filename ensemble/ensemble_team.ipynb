```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data1 = pd.read_csv("./output_anal/output_an_9265.csv")
data2 = pd.read_csv("./output_anal/output_h_9350.csv")
data3 = pd.read_csv("./output_anal/output_h_9355.csv")
data4 = pd.read_csv("./output_anal/output_h_9363.csv")
data5 = pd.read_csv("./output_anal/output_y_9222.csv")
```

```python
# 0부터 5까지 0.2 단위로 나누기, 앞뒤에 무한대 포함
bins = [-float('inf')] + [i / 10 for i in range(1, 51)] + [float('inf')]

# 레이블 설정, bins보다 하나 적은 수의 레이블 생성
labels = ['< -0.1'] + [f"{bins[i]:.1f} - {bins[i+1]:.1f}" for i in range(1, len(bins)-2)] + ['> 5.0']

# target 값을 기준으로 그룹화
data1['target_bins'] = pd.cut(data1['target'], bins=bins, labels=labels, right=False)

# 그룹화된 데이터프레임 생성
grouped_df = data1.groupby('target_bins').size().reset_index(name='count')

# 바 플롯 그리기
plt.figure(figsize=(14, 6))
sns.barplot(x='target_bins', y='count', data=grouped_df, palette='viridis')

# 제목과 레이블 설정
plt.title('Distribution of Target Values by 0.2 Score Bins', fontsize=16)
plt.xlabel('Target Score Bins', fontsize=14)
plt.ylabel('Count', fontsize=14)

# 플롯 보여주기
plt.xticks(rotation=45)  # x축 레이블 회전
plt.tight_layout()
plt.show()
```
