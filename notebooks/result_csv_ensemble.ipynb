```python
import pandas as pd
```

```python
data1 = pd.read_csv("data/ensemble/18_onlyswap.csv")
data1.head(10)
```

```python
data2 = pd.read_csv("data/ensemble/jaeryoung.csv")
data2.head(10)
```

```python
data3 = pd.read_csv("data/ensemble/deberta.csv")
data3.head(10)
```

```python
# Create an ensemble dataframe by averaging the 'target' columns from data1, data2, and data3
ensemble_df = pd.DataFrame(
    {
        "id": data1["id"],
        "target": (
            data1["target"] * 0.2 + data2["target"] * 0.6 + data3["target"] * 0.2
        ),
    }
)

ensemble_df.describe()
```

```python
ensemble_df["target"] = ensemble_df["target"]
ensemble_df.head()
```

```python
ensemble_df.to_csv("data/inference/ensemble-three.csv", index=False)
```
